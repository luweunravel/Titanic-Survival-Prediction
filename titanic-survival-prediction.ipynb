**Importing necessary libraries:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset from the Titanic-Dataset.csv file.**

```python
df = pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
```

**Exploring the dataset:**

```python
print("Dataset shape:", df.shape)
print("Dataset columns:", df.columns)
print("Dataset info:")
print(df.info())
print("Dataset summary statistics:")
print(df.describe())
```

**Data Visualization:**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Rates')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Filter the data to show passengers who survived
survivors = df[df['Survived'] == 1][['Name', 'Age', 'Sex', 'Survived']]
print("Passengers who survived:")
print(survivors)
```

```python
# Filter the data to show passengers who did not survive
non_survivors = df[df['Survived'] == 0][['Name', 'Age', 'Sex', 'Survived']]
print("Passengers who did not survive:")
print(non_survivors)
```

```python
# Relationship between age and survival rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Survived', data=df)
plt.title('Age vs Survival Rate')
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show()
```

```python
# Relationship between sex and survival rate
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Sex vs Survival Rate')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship between class and survival rate
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Class vs Survival Rate')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart to show the number of survivors and non-survivors
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df, hue='Survived')
plt.title('Survival Rates')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Histogram to show the distribution of ages
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
